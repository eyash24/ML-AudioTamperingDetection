```python
import random
import os
from pydub import AudioSegment

# deletion tampering
# pending: CSV Connection establishment

def deletionTampering(audio, clip_duration=10):
    durr = len(audio)
    print("Duration of audio:", durr)
    
    if durr < clip_duration * 1000:
        print("Warning: Audio is shorter than clip duration. Returning original audio.")
        return audio

    random_del = random.randint(1500, 2500)
    finish = random.randint(clip_duration * 1000 + random_del, durr)
    start = max(0, finish - clip_duration * 1000 - random_del)
    new_durr = clip_duration * 1000 + random_del
    random_cut = random.randint(300, new_durr)

    print("Start, Finish:", start, finish)
    print("Random Del, New Durr, Random Cut:", random_del, new_durr, random_cut)

    beginning = audio[start:start+random_cut]
    print("Length of Beginning:", len(beginning))
    print("beginning:", start, start+random_cut)

    end = audio[start+random_cut+random_del-100:finish]
    print("Length of End:", len(end))
    print("end:", start+random_cut+random_del-100, finish)

    del_audio = beginning + end
    print("Length of del_audio:", len(del_audio))

    if len(del_audio) > clip_duration * 1000:
        del_audio = del_audio[:clip_duration * 1000]
        print("Trimmed Length of del_audio:", len(del_audio))

    return del_audio

if __name__ == "__main__":
    audio_file_list = os.listdir(AUDIO_TRIAL_DIR)  # Ensure this is defined

    os.makedirs(TAMPERED_DIR, exist_ok=True)

    if not audio_file_list:
        print("No audio files found in directory.")
        exit()

    file_num = random.randint(0, len(audio_file_list) - 1)
    print("Selected File Index:", file_num)

    file_name = audio_file_list[file_num]
    file = os.path.join(AUDIO_TRIAL_DIR, file_name)
    
    print("Processing File:", file_name)

    org_audio = AudioSegment.from_file(file)
    print("Audio Type:", type(org_audio))

    durr = len(org_audio)
    print("Audio Duration:", durr)

    del_tampered_audio = deletionTampering(org_audio)
    print("Tampered Audio Type:", type(del_tampered_audio))

    del_file_name = f"{file_name[:-4]}_del.mp3"
    del_file_path = os.path.join(TAMPERED_DIR, del_file_name)
    del_tampered_audio.export(del_file_path, format="mp3")
```

```python
# insertion 
# testing of the module {remaining}
def insertionTampering(audio, clip_duration=10, lower_limit = 1000, higher_limit=1500):
    durr = len(audio)
    finish = random.randint(clip_duration*1000 + 100, durr)
    start = finish - clip_duration*1000

    audio_clip = audio[start:finish]

    while True:
        # Utilising UrbanSound8K dataset for noise
        fold = random.randint(1,10)
        fold_name = f"fold{fold}"
        fold_dir = os.path.join(NOISE_DIR,fold_name)
        fold_list = os.listdir(fold_dir)
        
        noise_n = random.randrange(0, len(fold_list))
        noise_file = fold_list[noise_n]
        noise_file = os.path.join(fold_dir, noise_file)
        noise = AudioSegment.from_wav(noise_file)

        # selecting audio noise clips that are greater than 1 second
        noise_durr = len(noise)
        if noise_durr >= lower_limit:
            break
        else:
            continue
    
    print("Noise file: ", noise_file)
    print("Noise duration: ", noise_durr)

    noise_durr = min(noise_durr, higher_limit)
    finish_noise = random.randint(1000, noise_durr)
    start_noise = 0

    add_noise = noise[start_noise:finish_noise]

    # Select a random cut point in the audio clip
    # Making sure that alteast 1000ms of noise is present in the original oudio clip
    random_cut = random.randint(0, clip_duration*1000-1000)
    
    beginning = audio_clip[:random_cut]
    end = audio_clip[random_cut:]

    insert_audio = beginning + add_noise + end
    print("len(insert_audio): ", len(insert_audio))
    print("beginning, random_cut, end: ", 0, random_cut, len(audio_clip))

    if len(insert_audio) > clip_duration*1000:
        insert_audio = insert_audio[:clip_duration*1000]
        print(len(insert_audio))
    
    return insert_audio


if __name__ == "__main__":
    file_num = random.randrange(1, len(audio_file_list))
    print(file_num)
    file_name = audio_file_list[file_num]
    file = os.path.join(AUDIO_TRIAL_DIR,file_name)
    comp_list = file_name[:-4].split("_")
    durr = int(comp_list[-1][:-1]) - int(comp_list[-3][:-1])
    print(file_name)

    org_audio = AudioSegment.from_file(file)
    print(type(org_audio))
    print(durr)
    insert_tampered_audio = insertionTampering(org_audio)
    print(type(insert_tampered_audio))
    insert_file_name = f"{file_name[:-4]}_insert.mp3"
    insert_file_path = os.path.join(TAMPERED_DIR, insert_file_name)
    insert_tampered_audio.export(insert_file_path, format="mp3")
```

```python
# Replacement
# testing of the module {remaining}
def replacementTampering(audio, clip_duration=10, lower_limit = 1000, higher_limit=2000):
    durr = len(audio)
    finish = random.randint(clip_duration*1000 + 100, durr)
    start = finish - clip_duration*1000

    audio_clip = audio[start:finish]

    # replace audio points 1
    r1_durr = random.randint(lower_limit, higher_limit)
    p2 = random.randint(higher_limit, durr)
    p1 = p2 - r1_durr

    while True:
        r2_durr = random.randint(lower_limit, higher_limit)
        p4 = random.randint(higher_limit, durr)
        if p4-r2_durr >= p2 or p4 <= p1:
            p3 = p4 - r2_durr
            break
    
    if p4 <= p2:
        p1,p2,p3,p4 = p3,p4,p1,p2

    c1 = audio_clip[:p1]
    c2 = audio_clip[p1:p2]
    c3 = audio_clip[p2:p3]
    c4 = audio_clip[p3:p4]
    c5 = audio_clip[p4:]

    # interchanging c2 and c3
    replace_audio = c1 + c4 + c3 + c2 + c5
    print("len(replace_audio): ", len(replace_audio))
    print("p1,p2,p3,p4: ", p1,p2,p3,p4)

    if len(replace_audio) > clip_duration*1000:
        replace_audio = replace_audio[:clip_duration*1000]
        print(len(replace_audio))
    
    return replace_audio
    

if __name__ == "__main__":
    file_num = random.randrange(1, len(audio_file_list))
    print(file_num)
    file_name = audio_file_list[file_num]
    file = os.path.join(AUDIO_TRIAL_DIR,file_name)
    comp_list = file_name[:-4].split("_")
    durr = int(comp_list[-1][:-1]) - int(comp_list[-3][:-1])
    print(file_name)

    org_audio = AudioSegment.from_file(file)
    print(type(org_audio))
    print(durr)
    replace_tampered_audio = replacementTampering(org_audio)
    print(type(replace_tampered_audio))
    replace_file_name = f"{file_name[:-4]}_replace.mp3"
    replace_file_path = os.path.join(TAMPERED_DIR, replace_file_name)
    replace_tampered_audio.export(replace_file_path, format="mp3")
```
